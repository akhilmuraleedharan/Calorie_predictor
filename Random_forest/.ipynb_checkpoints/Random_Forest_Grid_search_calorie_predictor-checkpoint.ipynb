### Import the required libraries

```python
import pandas as pd
import numpy as np

import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import metrics

from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV

import warnings
from warnings import filterwarnings
filterwarnings("ignore")
```

### Read the files and check the first few rows

```python
calories =  pd.read_csv("calories.csv")
exercise = pd.read_csv("exercise.csv")
```

```python
calories.head()
```

```python
exercise.head()
```

### Combine the two datasets

```python
data = pd.concat([exercise, calories["Calories"]],axis =1)
```

### Conduct some preliminary EDA by checking the datatypes and null values and duplicated values

```python
data.head()
```

```python
data.info()
```

```python
data.describe()
```

```python
data.isna().sum()
```

```python
data.duplicated().sum()
```

### Remove the User_ID column because of its irrelavancy

```python
data.drop(columns=["User_ID"],axis=1, inplace=True)
```

```python
data.head()
```

### There  is one categorical variable and it has to be converted to dummies

```python
data = pd.get_dummies(data, drop_first=True)
```

```python
plt.figure(figsize=(20,15))
plotnumber = 1

for column in data:
  if plotnumber <= 8:
    ax = plt.subplot(3,3,plotnumber)
    sns.histplot(data[column])
    plt.xlabel(column,fontsize=15)
  plotnumber+=1
plt.show()
```

### Check the heatmap to understand correlation

```python
plt.figure(figsize=(13,10))
sns.heatmap(data.corr(), cmap='Blues',annot = True)
plt.show()
```

### Divide the dataset as target variable and the independent variables

```python
X = data.drop(columns=["Calories"],axis=1)
y = data["Calories"]
```

```python
X.head()
```

```python
y.head()
```

### Split the data into test and train set

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.25, random_state=0)
```

### Conduct hyperparameter tuning to get the best estimator

```python
rf =  RandomForestRegressor()
cv_param = {'max_depth' : [2,4,None],
           'max_features': [0.3,0.6,1.0],'max_samples': [0.7],
           'min_samples_leaf' : [1,2],
           'min_samples_split' : [2,3],
            'n_estimators' : [75,100,125]}

rf_cv =  GridSearchCV(rf ,cv_param, scoring="r2", cv =5)
```

Fit the  model into the train data

```python
%%time
rf_cv.fit(X_train,y_train)
```

Obtain the best parameters

```python
rf_cv.best_params_
```

```python
y_pred =  rf_cv.best_estimator_.predict(X_test)
```

```python
print('r^2 score: ',metrics.r2_score(y_test,y_pred))

print('Mean Absolute Error:',metrics.mean_absolute_error(y_test,y_pred))
print('Mean Square Error:',metrics.mean_squared_error(y_test,y_pred))
print('Root Mean Square Error:',np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

### Check the feature importances of the independent variables

```python
importance =  rf_cv.best_estimator_.feature_importances_
rf_importance = pd.Series(importance, index= X_test.columns)
rf_importance = rf_importance.sort_values(ascending=False)

fig,ax =  plt.subplots(figsize=(8,6))
rf_importance.plot.bar(ax=ax)
```

```python
cdf = pd.DataFrame(rf_cv.best_estimator_.feature_importances_, X.columns, columns=['Coefficients'])
print(cdf)
```
